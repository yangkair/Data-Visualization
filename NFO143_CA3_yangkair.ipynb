```python
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
import numpy as np
pd.set_option("display.precision", 2)

mlp.__version__

import csv

data = {}
age = []
y = {}
o = []
n = []
i = 0
  
with open('census2000.csv','r') as csvfile:
     
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        t = row[1]+","+ row[2]
      
        if i != 0:
            if t not in y.keys():
                 y[t]  = 0
            y[t] += int(row[3])/10**6
            if row[2] not in age:
                age.append(row[2])
        i += 1
    i = 0
    data["Age"] =  age;
    for k in y.keys():
        if k[0:4] == "1900":
            o.append(y[k])
        else:
            n.append(y[k])
    data["1900"] = o
    data["2000"] = n

df = pd.DataFrame(data=data)

df.head()

```

```python
qual_colors=['#1f78b4','gold']

fig, ax = plt.subplots(figsize=(15,10)) 
bar_width = 0.4
x_pos = np.arange(len(df)) #bar location per category
#df.columns[]


for i, col in enumerate(df.columns[1:]):
    print(x_pos + i*bar_width)
    ax.bar( x_pos + i*bar_width, df[col], width=bar_width,  label=col, color=qual_colors[i])
    #ax.text(x_pos + i*bar_width, df[col]+4 , df[col],  ha='center')
   
for i in range(len(df)):

    ax.text(i + bar_width -0.5, data["1900"][i]+0.3, round(data["1900"][i],2), ha='center' )
for i in range(len(df)):

    ax.text(i + bar_width, data["2000"][i]+0.3, round(data["2000"][i],2), ha='center' )    

ax.set_ylabel('Population(Million)', fontsize=14)
ax.set_xlabel('Ages(Years Old)', fontsize=14)

ax.set_xticks(x_pos + bar_width)
ax.set_xticklabels(df['Age'], fontsize=14)
# remove chart border
for spine in plt.gca().spines.values():
    spine.set_visible(False)

#adding the grid lines for x-axis
ax.grid(visible=True, which='major', axis='y', color='#BFBEBE', lw=0.3) 
ax.set_title('Question: How has population changed in diferent age group from 19000 to 2000?\n United States\' Population has significantly increased in all ages from 1900 to 2000', fontsize=26, loc='Center', fontweight='bold' )

ax.legend()
```
